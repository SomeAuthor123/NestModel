```python
import networkx as nx
```

```python
edges = [(0,1), (1,2), (2,0), (3,4), (4,5)]
```

```python
G = nx.DiGraph(edges)
```

```python
nx.draw(G)
```

```python
nx.pagerank(G, tol=1e-20, max_iter=1000)
```

```python
edges2 = [(0,1), (1,2), (2,0), (3,5), (4,5)]
G2 = nx.DiGraph(edges2)
nx.pagerank(G2, tol=1e-20)
```

```python
nx.draw(G2)
```
